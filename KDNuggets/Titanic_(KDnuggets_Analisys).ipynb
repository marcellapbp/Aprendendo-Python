###Percentual de Mulheres que Sobreviveram 

```
#filtra sexo feminino e traz coluna de sobreviventes
woman = df[df['Sex'] =='female']['Survived']
#sum soma onde Survived == 1
#len traz total das linhas
perc = sum(woman)/len(woman) *100
```

```
#resultado 74% de todas as mulheres sobreviveram
print('Soma: ', sum(woman))
print('Total: ', len(woman))
print('Percentual: ', perc)
```

```
sobreviventes = df[df['Survived']==1]['Sex']
sobreviventes.head(15)
```

###Percentual de Sobreviventes que eram Mullheres e Homens

```
#de todos sobreviventes 68% deles eram mulheres e 31% homens
print('Sobreviventes Mulheres:', sum(sobreviventes == 'female'))
print('Sobreviventes Homens:', sum(sobreviventes == 'male'))
print('Total de Sobreviventes:', len(sobreviventes))
print('Percentual de Homens Sobreviventes', sum(sobreviventes == 'male')/len(sobreviventes) *100)
print('Percentual de Mulheres Sobreviventes', sum(sobreviventes == 'female')/len(sobreviventes) *100)
```
